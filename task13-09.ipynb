```python

import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd

driver = webdriver.Chrome()  

url = 'https://pay2igr.igrmaharashtra.gov.in/eDisplay/propertydetails'
driver.get(url)
```

```python
year_input = driver.find_element(By.ID, 'dbselect')
year_input.send_keys('2023')
```

```python
district_select = Select(driver.find_element(By.ID, 'district_id'))
district_select.select_by_visible_text('मुंबई उपनगर')
```

```python
taluka_select = Select(driver.find_element(By.ID, 'taluka_id'))
taluka_select.select_by_visible_text('अंधेरी')
```

```python
village_select = Select(driver.find_element(By.ID, 'village_id'))
village_select.select_by_visible_text('बांद्रा')
```

```python
year_input = driver.find_element(By.ID, 'free_text')
year_input.send_keys('2023')
```

```python
search_button = driver.find_element(By.ID, 'submit')
search_button.click()
```

```python
time.sleep(5)  
record_details_select = Select(driver.find_element(By.NAME, 'tableparty_length'))
record_details_select.select_by_value('50')
```

```python
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')

table = soup.find('table', {'id': 'tableparty'})
```

```python
s1 = table.text
```

```python
s1
```

```python
def clean_scrap_data(scraped_data):
    list1=scraped_data.split("List no. 2")
    empty=[]
    for i in list1:
        x1=i.split("\n")
        empty.append(x1)
    return empty   
    
```

```python
empty=clean_scrap_data(s1)
```

```python
empty_new=empty[1:-1]
empty_new
```

```python


# Use nested list comprehensions to remove empty strings and strings containing only spaces
filtered_data = [[item for item in sublist if item.strip() != ''] for sublist in empty_new]
```

```python
filtered_data
```

```python
df1=pd.DataFrame(filtered_data)
```

```python
df1
```

```python
dict1={0:'Sr_no',1:'Diarrhea_no',2:'Diarrhea_type',3:'Du_Prohibit_Office',4:'Year',5:'Buyer',6:'Seller',7:'Other_information'}
```

```python
df1.rename(columns=dict1, inplace=True)
```

```python
indexes_with_missing_values = df1[df1['Other_information'].isna()].index
list_index=indexes_with_missing_values.tolist()
list_index
```

```python
for i in list_index:
    data=df1["Seller"].loc[i]
    df1["Seller"].loc[i]=None
    df1['Other_information'].loc[i]=data
```
